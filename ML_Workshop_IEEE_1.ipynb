```python
import matplotlib.pyplot as plt
import numpy as np
```

Regression graph:

```python
points=np.linspace(1,500,100)

plt.plot(points)
plt.title("Data Points")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
randompoints=[0 for _ in range(100)]
for i in range(len(points)):
    randompoints[i]=points[i]+np.random.randint(-100,100)

plt.scatter(points,randompoints)
plt.title("Data Points with Randoms")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
plt.plot(points,points,'r')
plt.scatter(points,randompoints)
plt.title("Data Points")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.legend(['Best fit line'])
plt.show()
```

Find MSE,RMSE, Accuracy, MAE:

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
print("MAE:")
print(mean_absolute_error(points,randompoints))
print("MSE:")
print(mean_squared_error(points,randompoints))
print("RMSE:")
print(mean_squared_error(points,randompoints,squared=False))
```

Lets try to fit a Linear Regression Model to the data!

First, some preprocessing:

```python
randompoints=np.array(randompoints).reshape(-1,1)
points=np.array(points).reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(randompoints,points)
```

```python
model.score(randompoints,points)
```

```python
preds=model.predict(points)
print("MAE:")
print(mean_absolute_error(points,preds))
print("MSE:")
print(mean_squared_error(points,preds))
print("RMSE:")
print(mean_squared_error(points,preds,squared=False))
```

```python
plt.scatter(points,randompoints)
plt.plot(points,preds,'r')
plt.legend(['Prediction','Our Data'])
plt.show()
```
